```python
# you can also create a function to overcome the use of lamda function
df['spam'] = df['Category'].apply(lambda x: 1 if x =='spam' else 0)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.Message, df.spam, test_size=0.2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

v = CountVectorizer()

X_train_cv = v.fit_transform(X_train.values)
X_train_cv
```

Train the naive bayes model

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(X_train_cv, y_train)
```

```python
X_test_cv = v.transform(X_test)
```

```python
# whenever you have imbalance dataset you need to see the classification report 
# here we had very less number of span in comparison to hame
from sklearn.metrics import classification_report
y_pred = model.predict(X_test_cv)
print(classification_report(y_test, y_pred))
```

```python
emails = [
    'Hey mohan, can we get together to watch footbal game tomorrow?',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!'
]

emails_count = v.transform(emails)
model.predict(emails_count)
```

Train the model using sklearn pipeline and reduce number of lines of code

```python
from sklearn.pipeline import Pipeline

clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```
